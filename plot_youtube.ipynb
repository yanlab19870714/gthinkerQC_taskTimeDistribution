# Load Youtube Data Task Statistics

```python
import pandas as pd

df = pd.read_csv("youtube.csv")
df.head(5)
```

# Sort by Time

```python
df.sort_values("time(ms)", inplace=True)
df.tail(10)
```

# |V| v.s. Time Plot

```python
import numpy as np

x = np.array(df['|V|'])
y = np.array(df['time(ms)'])
```

### Ratio of {Total Time by All Tasks} / {Time of the Most Expensive Task}

```python
np.sum(y)/np.max(y)
```

### Kick out the last point which is an outlier that is on the far upper-right corner

```python
x = x[:-1]
y = y[:-1]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Number of Vertices in Task-Subgraph', fontsize=25)
plt.ylabel('Task-Subgraph Mining Time (ms)', fontsize=25)

plt.scatter(x, y, s=16)

plt.savefig('V.pdf')
plt.show()
```
