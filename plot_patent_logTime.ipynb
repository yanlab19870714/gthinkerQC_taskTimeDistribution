# Load Patent Data Task Statistics

```python
import pandas as pd

df = pd.read_csv("patent.csv")
df.head(5)
```

# Sort by Time

```python
df.sort_values("time(ms)", inplace=True)
df.tail(10)
```

# |V| v.s. Log-Time Plot

```python
import numpy as np

x = np.array(df['|V|'])
y = np.array(df['time(ms)'])
```

### Ratio of {Total Time by All Tasks} / {Time of the Most Expensive Task}

```python
np.sum(y)/np.max(y)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(20,10))

plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel('Number of Vertices in Task-Subgraph', fontsize=25)
plt.ylabel('Task-Subgraph Mining Time (ms)', fontsize=25)

plt.scatter(x, np.log(y), s=16)

plt.savefig('V3.pdf')
plt.show()
```
